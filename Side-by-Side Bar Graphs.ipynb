# Twinkle Twinkle Little Star

```python
import matplotlib.pyplot as plt

# Data
Original_TTLS = [4, 6, 6, 8, 8, 10]
Random_TTLS = [3, 9, 5, 10, 9, 6]
X_labels = ['A4', 'C4', 'D4', 'E4', 'F4', 'G4']

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.4
index = range(len(X_labels))

purple_bars = ax.bar(index, Original_TTLS, bar_width, color='slateblue', label='Original TTLS')
skyblue_bars = ax.bar([i + bar_width for i in index], Random_TTLS, bar_width, color='skyblue', label='Random TTLS')

# Adding labels
ax.set_xlabel("Note Used")
ax.set_ylabel("Frequency in Song")
ax.set_title("Frequency of Notes in \"Twinkle Twinkle Little Star\"")

ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(X_labels)
ax.legend()

# Counting and annotating each bar
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(purple_bars)
autolabel(skyblue_bars)

plt.tight_layout()
plt.show()
fig.savefig("TTLS Note Frequencies.png", bbox_inches='tight')
```
